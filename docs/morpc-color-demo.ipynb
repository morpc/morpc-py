```python
with open('../morpc/color/morpc_colors.json') as file:
    morpc_colors = json.load(file)
```

```python
morpc_colors['blue']
```

```python
all_hex=[]
for c in morpc_colors:
    c_hex=[]
    for pos in morpc_colors[c]['key']['qual']:
        hexs = morpc_colors[c]['gradient']['hex']
        c_hex.append(hexs[pos])
    all_hex.append(c_hex)
```

```python
all_hex
```

```python
import itertools
```

```python
morpc.plot_from_hex_list([x for x in itertools.chain.from_iterable([list for list in all_hex])])
```

```python
cmaps={}
for color in morpc_colors:
    cmaps[color] = morpc.get_continuous_cmap(morpc_colors[color]['gradient']['hex'])
```

```python
cmaps['darkgreen']
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)
data = np.random.randn(30, 30)
fig, ax = plt.subplots()
psm = ax.pcolormesh(data, cmap=cmaps['darkgreen'], rasterized=True, vmin=-4, vmax=4)
fig.colorbar(psm, ax=ax)
```

```python
from daltonize import daltonize
daltonize.simulate_mpl(fig, color_deficit='p')
```
